```python
from pypdf import PdfReader

read = PdfReader("/content/sugarcane.pdf")

text = ''

for page in read.pages:
    text += page.extract_text()

```

```python
!pip install -U langchain-text-splitters
```

```python
!pip install langchain
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size = 200,
    chunk_overlap = 50


)

chuks = splitter.split_text(text)
chuks
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("BAAI/bge-base-en-v1.5")

emmbeding = model.encode(chuks)
```

```python
!pip install faiss-cpu
```

```python
import faiss
import numpy as np

embeding = np.array(emmbeding).astype("float32")

index= faiss.IndexFlatL2(embeding.shape[1])

index.add(embeding)

```

```python
query = model.encode(
    ["in india how vrities "]
)

query = np.array(query).astype("float32")
```

```python
distance, indices = index.search(query, k=4)
```

```python
indices
```

```python
context = []

for i in indices[0]:
    context.append(chuks[i])
```
